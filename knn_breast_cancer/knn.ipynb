```python
import numpy as np
from sklearn import preprocessing, model_selection, neighbors
import pandas as pd

df =pd.read_csv('breast-cancer-wisconsin.data', index_col=0)
df.head()
```

```python
#df.set_index('clump_thickness')
print(df.dtypes)

#df.drop('id', axis=1, inplace=True)

df.replace('?', -99999, inplace=True)

df['bare_nuclei'] =df['bare_nuclei'].astype(np.int64)


print(df.dtypes)
df.head()
```

```python
X =np.array(df.drop('class', axis=True))
y =np.array(df['class'])

X_train, X_test, y_train, y_test =model_selection.train_test_split(X, y, test_size= 0.2)

clf =neighbors.KNeighborsClassifier()
clf.fit(X_train, y_train)

accuracy =clf.score(X_test, y_test)
print(accuracy)
```

```python
eg_set =np.array([4,2,1,1,1,2,3,2,1])
#    eg_set =eg_set.reshape(1, -1)

predict =clf.predict(eg_set)
predict
```

```python
from sklearn import svm

clf_2 =svm.SVC()
clf_2.fit(X_train, y_train)

accuracy_2 =clf_2.score(X_test, y_test)
print(accuracy_2)
```
